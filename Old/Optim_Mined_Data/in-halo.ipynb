Halo properties found by _Rockstar_:

```python
f = open('lcdm_halos_0.ascii', "r", encoding="ascii")
df_hf = pd.read_csv(f, sep=' ', skiprows=range(1,19), skipinitialspace=True)
df_hf
```

Keeping only the centers' positions and the radii around the center in which half of the halo mass is distributed.

```python
dfh= df_hf[['x','y','z','Halfmass_Radius']]
dfh
```

```python
from tqdm import tqdm
```

Radius in which most of the mass of halo is located is about 20 times of the halfmass radius.
The radii are in unit $(\frac{kpc}{h})$ .
Positions are in unit $(\frac{Mpc}{h})$ .

```python
def in_or_out(s, hf):
    in_out = np.zeros(len(s))
    ss = s.to_numpy()
    hh = hf.to_numpy()
    for j in tqdm(range(len(hf))):
        r = hh[j,3] / 50
        x = hh[j,0]
        y = hh[j,1]
        z = hh[j,2]
        dis = np.sqrt(np.square(ss[:,1]/1000-x) + np.square(ss[:,2]/1000-y) + np.square(ss[:,3]/1000-z))
        in_out = in_out+(dis<=r)
    in_out[in_out>0]=1
    return in_out
```

```python
a=(in_or_out(df,dfh))
```

```python
ran.insert(1,'in-halo',a)
```
