```python
from pygadgetreader import *
import pandas as pd
import numpy as np
```

```python
vel = readsnap('lcdmML_snap000_cdm','vel','dm')
vx = vel[:,0].tolist()
vy = vel[:,1].tolist()
vz = vel[:,2].tolist()
#mass = readsnap('lcdmML_snap000_cdm','mass','dm').tolist()
particle_id = readsnap('lcdmML_snap000_cdm','pid','dm').tolist()
position = readsnap('lcdmML_snap000_cdm','pos','dm')
x = position[:,0].tolist()
y = position[:,1].tolist()
z = position[:,2].tolist()
```

```python
data = {'particle_id': particle_id,
        'x': x,
        'y': y,
        'z': z,
        'vx': vx,
        'vy': vy,
        'vz': vz}
df100 = pd.DataFrame(data)
```

```python
df100.insert(1, 'in-halo', 0)
```

```python
df100['in-halo'] = np.random.choice([0, 1], df100.shape[0])
df100.head()
```

```python
df = df100.sort_values( by =['x'])
```

```python
data = df.copy(deep = True)
data.x = data.x / 1000
data.y = df.y / 1000
data.z = df.z / 1000
print(data.x.max())
print(data.y.max())
print(data.z.max())
```

```python
box_size = 159
boundary = 20

d = data[data.x > boundary]

d = d[data.x < box_size - boundary]

d = d[data.y > boundary]

d = d[data.y < box_size - boundary]

d = d[data.z > boundary]

d = d[data.z < box_size - boundary]
```

```python
def box_mass(r_cutoff, df, d):
    m = len(d)
#     print(m)
    mass = pd.DataFrame(np.zeros(m),columns=[f'den_contrast({r_cutoff})'])
    for i in tqdm(range(m)):
        wx=d[i,0]
        wy=d[i,1]
        wz=d[i,2]
        
        temp = np.where((df[:,0] > wx - r_cutoff) & (df[:,0] < wx + r_cutoff) & (df[:,1] < wy+r_cutoff) & (df[:,1] > wy-r_cutoff) & (df[:,2] < wz+r_cutoff) & (df[:,2] > wz-r_cutoff) )
    
        mass.iloc[i,-1] = len(temp[0])
#         print(i)
    return mass
```

```python
n = 100

sample_keep = d.sample(n, random_state = 1)
sample = sample_keep[['x','y','z']]
sample_density = sample_keep[['particle_id', 'in-halo']]
sample_density.reset_index(drop = True, inplace = True)
sample=sample.to_numpy()
sample_density.head()
```

```python
data2 = data[['x','y','z']].to_numpy()
```

```python
rc = np.array([])
for i in range(8,11, 1):
    r = round(0.8 + 0.4 * i,2)
    rc = np.append(rc, r)
    density = box_mass(r , data2, sample)
    density/=((2*r)**3)
    sample_density = pd.concat([sample_density, density], axis = 1)
sample_density.head()
```

```python
rho_bar = len(df) / (159.8**3)
sample_density.iloc[:,2:] -= rho_bar
sample_density.iloc[:,2:] /= rho_bar
sample_density
```
